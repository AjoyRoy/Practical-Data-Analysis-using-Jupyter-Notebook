```python
import pandas as pd
```

```python
df_stock_price=pd.read_csv('AAPL.csv', index_col=0,parse_dates=True)
```

```python
df_stock_price.head(3)
```

```python
df_stock_price.index
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
df_stock_price['Close'].plot();
```

```python
df_stock_price['Close'].plot(linewidth=.5, color='green',linestyle='dotted');
```

```python
df_stock_price['Close'].plot(linewidth=.5, color='green',linestyle='dotted')
plt.xlabel('Close Date')
plt.ylabel('Close Price')
plt.suptitle('AAPL Stock Price Trend');
```

```python
plt.bar(df_stock_price.index,df_stock_price['Volume']);
```

```python
plt.bar(df_stock_price.index,df_stock_price['Volume'], color='blue')
plt.xticks(rotation = 90)
plt.yticks(fontsize = 10)
plt.xlabel('Date')
plt.ylabel('Volume')
plt.suptitle('AAPL Stock Volume Trend');
```
